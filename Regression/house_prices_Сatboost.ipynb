```python
import pandas as pd

from catboost import CatBoostRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
import time
```

```python
X_data = pd.read_csv("X_data_cat.csv")
y_data = pd.read_csv("y_data_cat.csv")
```

```python
X_data
```

```python
# Оставляем категориальные переменные
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
```

```python
model_cat = CatBoostRegressor(verbose=False, random_state=42)
```

```python
params = {
    'depth': [5, 7],
    'learning_rate': [0.1],
    'iterations': [100, 200],
    'l2_leaf_reg':[2, 3]}
```

```python
grid_search_cv_cat = GridSearchCV(model_cat, params, cv=5, n_jobs=-2)
```

```python
cat_features = ['MSSubClass', 'LotShape', 'LotConfig', 'Neighborhood', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'ExterQual', \
'Foundation', 'BsmtQual', 'BsmtExposure', 'BsmtFinType1', 'HeatingQC', 'KitchenQual', 'GarageType', 'GarageFinish']
```

```python
start_time = time.time()
grid_search_cv_cat.fit(X_train, y_train, cat_features=cat_features)
end_time = time.time()
time_fit = end_time - start_time
```

```python
# Время предыдущего обучения (для сравнения с GPU) - 60.3c
print(round(time_fit, 1))
```

```python
grid_search_cv_cat.best_params_
```

```python
best_model = grid_search_cv_cat.best_estimator_
```

```python
best_model.score(X_train, y_train)
```

```python
best_model.score(X_test, y_test)
```
